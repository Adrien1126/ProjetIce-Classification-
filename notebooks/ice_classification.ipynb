### Analyse des caractéristiques :
**Valeurs centrales** (Moyenne et Médiane) :

La variable t2m (température à 2 m sous la surface) a une moyenne de -10,19°C et une médiane de -9,6°C, ce qui suggère des températures globalement froides, comme attendu pour le Groenland.
Certaines caractéristiques comme SIC (Sea Ice Concentration) sont très concentrées autour de valeurs élevées (médiane = 84,6%), ce qui reflète une forte couverture de glace.

**Dispersion** (Écart-type et Plages) :

La caractéristique r1_MAR a une moyenne de 18,78 mais une plage très large (min = 0, max = 479,72), ce qui pourrait indiquer des valeurs aberrantes ou une grande variabilité régionale.
Les caractéristiques comme u10 et v10 (composantes du vent) sont proches de 0 en moyenne, mais leur écart-type montre une variabilité modérée.

**Données potentiellement inutiles** :

Certaines caractéristiques comme r3_MAR, r4_MAR, et r5_MAR ont des médianes égales à 0 et des valeurs très faibles dans leurs 25e et 75e percentiles, ce qui pourrait suggérer qu'elles apportent peu d'information pour la classification. Cela devra être validé par des tests de pertinence des caractéristiques.

**Outliers possibles** :

Les valeurs maximales de r1_MAR, r2_MAR, et SST sont très élevées par rapport aux 75e percentiles. Ces outliers pourraient influencer les modèles si non traités.

### Analyse des cibles :
**Problème de sparsité** :

Toutes les cibles $(Y_1,Y_3,Y_4,Y_5)$ ont une médiane et des percentiles (25 %, 50 %, 75 %) égaux à 0. Cela signifie qu'une grande majorité des valeurs des cibles sont nulles.
Par exemple, pour $Y_1$, seulement quelques valeurs dépassent 0, mais la moyenne (3,52) et l'écart-type (18,98) montrent qu'il existe quelques valeurs très élevées (max = 433).

**Imbalance** (Déséquilibre des classes) :

Une fois $Y_1$ transformé en variable binaire, les classes pourraient être déséquilibrées (beaucoup de 0 et peu de 1). Cela nécessitera potentiellement des techniques de gestion du déséquilibre (comme la pondération des classes ou le suréchantillonnage).

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import matplotlib.pyplot as plt

# Préparation des données (caractéristiques et cible)
X = numerical_features.drop(columns=['Y1_binary'])  # Exclure la cible
y = numerical_features['Y1_binary']  # Cible

# Entraîner un modèle Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X, y)

# Importances des caractéristiques
feature_importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

# Afficher les importances
print(feature_importances)

# Tracer les importances des caractéristiques
plt.figure(figsize=(10, 6))
plt.barh(feature_importances['Feature'], feature_importances['Importance'], color='skyblue')
plt.title("Importance des Caractéristiques (Random Forest)")
plt.xlabel("Importance")
plt.ylabel("Caractéristiques")
plt.gca().invert_yaxis()  # Inverser l'ordre pour afficher les plus importantes en haut
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

# Définir des combinaisons de caractéristiques
feature_sets = {
    'All Features': X,
    'MAR_PCA + t2m': X[['MAR_PCA', 't2m']],
    'Top 3 Features': X[['MAR_PCA', 't2m', 'u10']]
}

# Tester chaque ensemble de caractéristiques
for name, features_subset in feature_sets.items():
    model = RandomForestClassifier(random_state=42)
    scores = cross_val_score(model, features_subset, y, cv=5, scoring='accuracy')
    print(f"{name}: Mean Accuracy = {scores.mean():.4f}")
```

```python
# Sélection des caractéristiques finales
X_final = X[['MAR_PCA', 't2m', 'u10']]
y_final = y  # Cible (Y1_binary)

# Division des données en ensembles d'entraînement et de test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, roc_curve

# Entraîner le modèle final Random Forest
final_rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
final_rf_model.fit(X_train, y_train)

# Prédictions
y_pred = final_rf_model.predict(X_test)
y_proba = final_rf_model.predict_proba(X_test)[:, 1]

# Évaluation des performances
print("Rapport de Classification - Modèle Final (Random Forest) :")
print(classification_report(y_test, y_pred))

# Calculer l'AUC-ROC
roc_auc = roc_auc_score(y_test, y_proba)
print(f"AUC-ROC: {roc_auc:.4f}")
```

```python
# Courbe ROC
fpr, tpr, _ = roc_curve(y_test, y_proba)

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"Random Forest (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--', label="Baseline")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Courbe ROC - Modèle Final")
plt.legend()
plt.show()
```

```python
# Obtenir les probabilités pour la classe positive (1)
y_proba = final_rf_model.predict_proba(X_test)[:, 1]  # Probabilités de la classe 1
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

# Liste des seuils à tester
thresholds = np.linspace(0.1, 0.9, 9)  # Seuils de 0.1 à 0.9
metrics = []

# Tester chaque seuil
for threshold in thresholds:
    # Prédictions binaires basées sur le seuil
    y_pred_threshold = (y_proba > threshold).astype(int)
    
    # Calcul des métriques
    precision = precision_score(y_test, y_pred_threshold)
    recall = recall_score(y_test, y_pred_threshold)
    f1 = f1_score(y_test, y_pred_threshold)
    
    # Stocker les résultats
    metrics.append({
        'Threshold': threshold,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1
    })
```

```python
import pandas as pd

# Convertir les résultats en DataFrame
metrics_df = pd.DataFrame(metrics)

# Afficher les résultats
print(metrics_df)

# Identifier le meilleur seuil selon une métrique (par exemple, F1-Score)
best_threshold = metrics_df.loc[metrics_df['F1-Score'].idxmax(), 'Threshold']
print(f"Meilleur seuil basé sur le F1-Score: {best_threshold:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(metrics_df['Threshold'], metrics_df['Precision'], label='Precision', marker='o')
plt.plot(metrics_df['Threshold'], metrics_df['Recall'], label='Recall', marker='o')
plt.plot(metrics_df['Threshold'], metrics_df['F1-Score'], label='F1-Score', marker='o')
plt.axvline(best_threshold, color='red', linestyle='--', label=f"Best Threshold: {best_threshold:.2f}")
plt.title("Évolution des métriques en fonction du seuil")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.legend()
plt.grid()
plt.show()
```

```python
# Prédictions avec le seuil optimal
optimal_threshold = 0.40
y_pred_optimal = (y_proba > optimal_threshold).astype(int)

# Évaluer les performances avec le seuil optimal
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_optimal))
```

## Étape 1 : Validation Croisée pour Estimer la Robustesse
Sans ensemble de test indépendant, vous pouvez utiliser la validation croisée pour évaluer la robustesse et la généralisation de votre modèle.

Implémenter une Validation Croisée

Utilisez une validation croisée stratifiée pour préserver le déséquilibre des classes.

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import make_scorer, f1_score

# Définir une validation croisée stratifiée
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Utiliser le F1-score comme métrique
f1_scorer = make_scorer(f1_score)

# Évaluer le modèle avec validation croisée
cv_scores = cross_val_score(final_rf_model, X_final, y_final, cv=skf, scoring=f1_scorer)

print(f"Scores F1 pour chaque fold : {cv_scores}")
print(f"F1-score moyen : {cv_scores.mean():.4f}")
```

Si ce score vous semble suffisant, vous pouvez le considérer comme un résultat final. Sinon, testez un modèle avancé comme XGBoost ou LightGBM pour voir si vous pouvez améliorer les performances.


```python
from xgboost import XGBClassifier

# Entraîner XGBoost avec un poids pour gérer le déséquilibre
xgb_model = XGBClassifier(random_state=42, scale_pos_weight=len(y_final[y_final == 0]) / len(y_final[y_final == 1]))
xgb_scores = cross_val_score(xgb_model, X_final, y_final, cv=skf, scoring=f1_scorer)

print(f"Scores F1 pour XGBoost : {xgb_scores}")
print(f"F1-score moyen pour XGBoost : {xgb_scores.mean():.4f}")

```

2. Optimisation des Hyperparamètres

Vous pouvez optimiser les hyperparamètres de votre modèle Random Forest pour maximiser ses performances.

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid_xgb = {
    'learning_rate': [0.1, 0.2],
    'max_depth': [3, 5],
    'n_estimators': [50, 100],
    'scale_pos_weight': [len(y_final[y_final == 0]) / len(y_final[y_final == 1])]
}


random_search_xgb = RandomizedSearchCV(
    estimator=XGBClassifier(random_state=42, eval_metric='logloss'),
    param_distributions=param_grid_xgb,
    n_iter=10,  # Nombre de combinaisons à tester
    cv=5,
    scoring='f1_macro',
    random_state=42
)
random_search_xgb.fit(X_final, y_final)

print("Meilleurs paramètres :", random_search_xgb.best_params_)
```

```python
# Réentraîner XGBoost avec les meilleurs paramètres
best_xgb_model = XGBClassifier(
    random_state=42,
    eval_metric='logloss',
    scale_pos_weight=7.69,
    n_estimators=50,
    max_depth=3,
    learning_rate=0.1
)
best_xgb_model.fit(X_final, y_final)

# Validation croisée avec les meilleurs paramètres
best_xgb_scores = cross_val_score(best_xgb_model, X_final, y_final, cv=skf, scoring=f1_scorer)
print(f"Scores F1 après optimisation pour XGBoost : {best_xgb_scores}")
print(f"F1-score moyen après optimisation : {best_xgb_scores.mean():.4f}")
```
